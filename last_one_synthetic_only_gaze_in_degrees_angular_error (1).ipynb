```python
# ============================================================
# Synthetic-only Fine-tuning (ResNet50, init from REAL-trained last.pt)
#  - Data: Blender synthetic (frames + Euler -> TEyeD-style gaze vector)
#  - Init: last.pt (REAL ile eÄŸitilmiÅŸ ResNet50 checkpoint'i)
#  - Train: Sadece SYNTHETIC (REAL yok)
#  - Loss: 0.7 * (1 - cos) + 0.3 * MSE (normalized vectors)
#  - LR: backbone 1e-4, head 5e-4 + 5-epoch warmup + ReduceLROnPlateau
#  - Output: best_gaze_synthetic_from_real_lastpt_resnet50.pt
# ============================================================

import os, json, random, numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms as T, models
from tqdm import tqdm
import matplotlib.pyplot as plt

# ---------------- Paths ----------------
ROOT_DIR  = "/content/drive/MyDrive/dataset_split/synthetic_data/left"
CKPT_PATH = "/content/drive/MyDrive/dataset_split/last.pt"  # BU REAL Ä°LE EÄžÄ°TÄ°LMÄ°Åž MODEL
SAVE_BEST = "/content/drive/MyDrive/dataset_split/best_gaze_synthetic_from_real_lastpt_resnet50.pt"

# ---------------- Env ----------------
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)
torch.manual_seed(42)
if device == "cuda":
    torch.cuda.manual_seed_all(42)
np.random.seed(42)
random.seed(42)
EPS = 1e-8

# ============================================================
# Euler (x,y,z) -> rotation matrix -> 3D gaze vector (TEyeD-style)
# ============================================================

def euler_xyz_to_matrix(x, y, z):
    """
    x, y, z: Euler aÃ§Ä±larÄ± (radyan), XYZ sÄ±rasÄ±
    R = Rz * Ry * Rx
    """
    cx, cy, cz = np.cos(x), np.cos(y), np.cos(z)
    sx, sy, sz = np.sin(x), np.sin(y), np.sin(z)

    Rx = np.array([
        [1,  0,   0],
        [0, cx, -sx],
        [0, sx,  cx]
    ])

    Ry = np.array([
        [ cy, 0, sy],
        [  0, 1,  0],
        [-sy, 0, cy]
    ])

    Rz = np.array([
        [cz, -sz, 0],
        [sz,  cz, 0],
        [ 0,   0, 1]
    ])

    return Rz @ Ry @ Rx

def new_euler_to_gaze_vector(x, y, z):
    """
    Blender XYZ Euler -> local forward (-Z) -> TEyeD koordinat sistemi:
      g_blender = R * [0, 0, -1]
      g_teyed   = [ g.x, -g.y, -g.z ]
    SonrasÄ±nda normalize.
    """
    R = euler_xyz_to_matrix(x, y, z)
    d_local = np.array([0.0, 0.0, -1.0], dtype=np.float32)  # forward in Blender
    g = R @ d_local
    g = np.array([g[0], -g[1], -g[2]], dtype=np.float32)    # TEyeD mapping
    n = np.linalg.norm(g)
    if n < 1e-8:
        return 0.0, 0.0, -1.0
    g /= n
    return float(g[0]), float(g[1]), float(g[2])

# ============================================================
# Dataset (GAZE VECTORS) - no mean/std
# ============================================================

class RGBOnlyDataset(Dataset):
    """
    Synthetic:
      ROOT_DIR/split/render_X/
        frame_001.png .. frame_125.png
        render_metadata.json (rotation_over_time: {frame: {x,y,z}})
    """
    def __init__(self, root_dir, split='train', transform=None):
        self.root_dir = os.path.join(root_dir, split)
        self.transform = transform
        self.samples = []   # (frame_path, (gx,gy,gz))

        if not os.path.exists(self.root_dir):
            print(f"UYARI: Veri seti dizini bulunamadÄ±: {self.root_dir}")
            return

        for fld in sorted(os.listdir(self.root_dir)):
            render_path = os.path.join(self.root_dir, fld)
            js = os.path.join(render_path, "render_metadata.json")
            if not os.path.exists(js):
                continue
            try:
                with open(js, "r") as f:
                    meta = json.load(f)
                rot = meta.get("rotation_over_time", {})
                for i in range(1, 126):
                    fp = os.path.join(render_path, f"frame_{i:03d}.png")
                    if not os.path.exists(fp) or str(i) not in rot:
                        continue
                    r = rot[str(i)]
                    if not all(k in r for k in ("x","y","z")):
                        continue

                    ex = float(r["x"])
                    ey = float(r["y"])
                    ez = float(r["z"])

                    # Euler -> gaze vector (TEyeD-style)
                    gx, gy, gz = new_euler_to_gaze_vector(ex, ey, ez)

                    self.samples.append((fp, (gx, gy, gz)))
            except Exception as e:
                print(f"Parse error processing {render_path}: {e}")

        print(f"{split}: {len(self.samples)} samples")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        fp, gaze = self.samples[idx]
        img = Image.open(fp).convert("RGB")
        if self.transform:
            img = self.transform(img)
        tgt = torch.tensor(gaze, dtype=torch.float32)  # (gx, gy, gz)
        return img, tgt

# ============================================================
# Transforms
#  - Train: augmentation + normalize
#  - Val/Test: sadece resize + normalize
# ============================================================

IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]

train_tf = T.Compose([
    T.Grayscale(num_output_channels=3),  # ResNet50 expects 3 channels
    T.Resize((224, 224)),
    T.ColorJitter(brightness=0.2, contrast=0.2),
    T.ToTensor(),
    T.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

eval_tf = T.Compose([
    T.Grayscale(num_output_channels=3),
    T.Resize((224, 224)),
    T.ToTensor(),
    T.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

# ---------------- Data ----------------
train_dataset = RGBOnlyDataset(ROOT_DIR, 'train', train_tf)
val_dataset   = RGBOnlyDataset(ROOT_DIR, 'val',   eval_tf)
test_dataset  = RGBOnlyDataset(ROOT_DIR, 'test',  eval_tf)

BATCH_SIZE = 8
NUM_WORKERS = 4 if device=="cuda" else 2

def make_loader(ds, shuffle):
    return DataLoader(
        ds,
        batch_size=BATCH_SIZE,
        shuffle=shuffle,
        num_workers=NUM_WORKERS,
        pin_memory=(device=="cuda"),
        persistent_workers=(device=="cuda" and NUM_WORKERS>0 and len(ds)>0),
        prefetch_factor=4 if device=="cuda" else 2,
    )

train_loader = make_loader(train_dataset, True)
val_loader   = make_loader(val_dataset,   False)
test_loader  = make_loader(test_dataset,  False)
print(f"Train: {len(train_dataset)} | Val: {len(val_dataset)} | Test: {len(test_dataset)}")

# ============================================================
# Model
# ============================================================

gaze_model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
gaze_model.fc = nn.Linear(gaze_model.fc.in_features, 3)  # 3D gaze vector
gaze_model = gaze_model.to(device)

# ---------------- Load REAL-trained last.pt as init ----------------
try:
    ckpt = torch.load(CKPT_PATH, map_location="cpu", weights_only=False)
    sd = ckpt.get("state_dict", ckpt)
    rep = gaze_model.load_state_dict(sd, strict=False)
    print("Partial load from REAL-trained last.pt done.")
    print(f"Missing: {len(rep.missing_keys)} | Unexpected: {len(rep.unexpected_keys)}")
except Exception as e:
    print(f"Checkpoint load skipped ({e}).")

# ============================================================
# Loss & Metrics (normalized vectors)
# ============================================================

def combined_loss(pred, tgt, alpha=0.7):
    """
    alpha * (1 - cos) + (1 - alpha) * MSE, normalize vektÃ¶rler Ã¼zerinde
    """
    pred_n = F.normalize(pred, dim=1)
    tgt_n  = F.normalize(tgt,  dim=1)
    cos = torch.sum(pred_n * tgt_n, dim=1).clamp(-1.0, 1.0)
    cos_loss = (1.0 - cos).mean()
    mse_loss = F.mse_loss(pred_n, tgt_n)
    return alpha * cos_loss + (1.0 - alpha) * mse_loss

def angular_error_deg(pred, gt):
    pred = F.normalize(pred, dim=1)
    gt   = F.normalize(gt,   dim=1)
    cos_sim = torch.clamp(torch.sum(pred * gt, dim=1), -1.0, 1.0)
    return torch.rad2deg(torch.acos(cos_sim))

# ============================================================
# Optimizer: differential LR + warmup + scheduler
# ============================================================

head_params = list(gaze_model.fc.parameters())
backbone_params = [p for n,p in gaze_model.named_parameters() if not n.startswith('fc.')]

optimizer = torch.optim.AdamW(
    [{'params': backbone_params, 'lr': 1e-4},
     {'params': head_params,     'lr': 5e-4}],
    weight_decay=1e-4
)

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', factor=0.5, patience=3
)

WARMUP_EPOCHS = 5
BASE_LR_BB, BASE_LR_HD = 1e-4, 5e-4

def apply_warmup(opt, epoch):
    if epoch <= WARMUP_EPOCHS:
        k = epoch / float(WARMUP_EPOCHS)  # 0.2 â†’ 1.0
        opt.param_groups[0]['lr'] = BASE_LR_BB * k
        opt.param_groups[1]['lr'] = BASE_LR_HD * k

# ============================================================
# Train
# ============================================================

EPOCHS = 50
best_val = float("inf")
train_losses, val_losses, val_angles, bb_lrs, hd_lrs = [], [], [], [], []

for epoch in range(1, EPOCHS+1):
    gaze_model.train()
    apply_warmup(optimizer, epoch)

    tr_loss = 0.0
    for x, t in tqdm(train_loader, desc=f"Epoch {epoch}/{EPOCHS} [Train]", leave=False):
        x, t = x.to(device), t.to(device)
        optimizer.zero_grad(set_to_none=True)
        out = gaze_model(x)
        loss = combined_loss(out, t, alpha=0.7)
        loss.backward()
        optimizer.step()
        tr_loss += loss.item()
    tr_loss /= max(1, len(train_loader))

    gaze_model.eval()
    va_loss, va_ang_list = 0.0, []
    with torch.no_grad():
        for x, t in tqdm(val_loader, desc=f"Epoch {epoch}/{EPOCHS} [Val]", leave=False):
            x, t = x.to(device), t.to(device)
            out = gaze_model(x)
            loss = combined_loss(out, t, alpha=0.7)
            va_loss += loss.item()
            va_ang_list.append(angular_error_deg(out, t).mean().item())
    va_loss /= max(1, len(val_loader))
    va_ang = float(np.mean(va_ang_list)) if len(va_ang_list) else float('inf')

    train_losses.append(tr_loss)
    val_losses.append(va_loss)
    val_angles.append(va_ang)
    bb_lrs.append(optimizer.param_groups[0]['lr'])
    hd_lrs.append(optimizer.param_groups[1]['lr'])

    print(f"Epoch {epoch:02d} | Train Loss: {tr_loss:.8f} | Val Loss: {va_loss:.8f} "
          f"| Val Angular: {va_ang:.2f}Â° | LR(bb,hd)=({bb_lrs[-1]:.8f},{hd_lrs[-1]:.8f})")

    prev_bb, prev_hd = optimizer.param_groups[0]['lr'], optimizer.param_groups[1]['lr']
    scheduler.step(va_loss)
    if optimizer.param_groups[0]['lr'] < prev_bb or optimizer.param_groups[1]['lr'] < prev_hd:
        print(f"ðŸ”» LR reduced to backbone={optimizer.param_groups[0]['lr']:.8f}, head={optimizer.param_groups[1]['lr']:.8f}")

    if va_loss + 1e-12 < best_val:
        best_val = va_loss
        torch.save(gaze_model.state_dict(), SAVE_BEST)
        print("âœ… Best synthetic-only model saved:", SAVE_BEST)

print("Best Val Loss:", best_val)

# ============================================================
# Test
# ============================================================

gaze_model.load_state_dict(torch.load(SAVE_BEST, map_location=device))
gaze_model.to(device)
gaze_model.eval()
test_angles = []
with torch.no_grad():
    for x, t in tqdm(test_loader, desc="Test"):
        x, t = x.to(device), t.to(device)
        ang = angular_error_deg(gaze_model(x), t)
        test_angles.append(ang.mean().item())
print(f"âœ… SYNTHETIC-ONLY TEST Angular Error: {np.mean(test_angles):.2f}Â°")

# ============================================================
# Plots
# ============================================================

fig, ax1 = plt.subplots(figsize=(9,5))
ax1.plot(train_losses, label="Train Loss")
ax1.plot(val_losses, label="Val Loss")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.grid(True)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.plot(val_angles, label="Val Angular (Â°)", linestyle='--')
ax2.set_ylabel("Angular Error (Â°)")
ax2.legend(loc='upper right')

plt.title("Training Progress (Synthetic-only, init from REAL last.pt)")
plt.show()

plt.figure(figsize=(9,4))
plt.plot(bb_lrs, label="LR Backbone")
plt.plot(hd_lrs, label="LR Head")
plt.title("Learning Rate Schedule")
plt.xlabel("Epoch")
plt.ylabel("LR")
plt.grid(True)
plt.legend()
plt.show()
```
