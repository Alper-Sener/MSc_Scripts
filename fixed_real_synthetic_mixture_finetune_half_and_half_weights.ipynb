```python
# ============================================================
# Mixed Fine-tuning (Synthetic + REAL TEyeD)
#  - Model: ResNet50
#  - Start from synthetic checkpoint last.pt
#  - Datasets:
#      * Synthetic: RGBOnlyDataset (render frames + Euler -> gaze vec)
#      * Real: TEyeDRealDataset (mp4 + mp4gaze_vec.txt, FRAME;X;Y;Z;)
#  - Loss: combined_loss = 0.7 * (1 - cos) + 0.3 * MSE  (normalized vectors)
#  - Training: loss = 0.9 * loss_real + 0.1 * loss_synth (opsiyonel, synth yoksa 1.0 * real)
#  - LR: low (backbone 1e-5, head 5e-5) + 5-epoch warmup + ReduceLROnPlateau
#  - Validation/Test: REAL TEyeD only (ana hedef)
# ============================================================

import os, json, random, numpy as np, cv2
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms as T, models
from tqdm import tqdm
import matplotlib.pyplot as plt

# ---------------- Paths ----------------
SYNTH_ROOT = "/content/drive/MyDrive/dataset_split/synthetic_data/left"
REAL_ROOT  = "/content/drive/MyDrive/dataset_split/real_gaze_vectors_videos_and_texts_split"

CKPT_PATH  = "/content/drive/MyDrive/dataset_split/last.pt"  # pretrain checkpoint
SAVE_BEST  = "/content/drive/MyDrive/dataset_split/best_gaze_finetuned_mixed_synth_real.pt"

# ---------------- Env ----------------
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)
torch.manual_seed(42)
if device == "cuda":
    torch.cuda.manual_seed_all(42)
np.random.seed(42)
random.seed(42)
EPS = 1e-8

# ============================================================
# 1) Euler -> Gaze Vector (Synthetic, mapped to TEyeD-style)
# ============================================================

def euler_xyz_to_matrix(x, y, z):
    cx, cy, cz = np.cos(x), np.cos(y), np.cos(z)
    sx, sy, sz = np.sin(x), np.sin(y), np.sin(z)

    Rx = np.array([[1,0,0],
                   [0,cx,-sx],
                   [0,sx, cx]])

    Ry = np.array([[ cy,0,sy],
                   [  0,1, 0],
                   [-sy,0,cy]])

    Rz = np.array([[cz,-sz,0],
                   [sz, cz,0],
                   [ 0,  0,1]])

    return Rz @ Ry @ Rx

def new_euler_to_gaze_vector(x, y, z):
    """
    Blender XYZ Euler -> forward (-Z) -> TEyeD koordinat sistemi:
      g_blender = R * [0, 0, -1]
      g_teyed   = [ g.x, -g.y, -g.z ]
    Sonrasƒ±nda normalize.
    """
    R = euler_xyz_to_matrix(x, y, z)
    d_local = np.array([0.0, 0.0, -1.0], dtype=np.float32)
    g = R @ d_local
    g = np.array([ g[0], -g[1], -g[2] ], dtype=np.float32)
    n = np.linalg.norm(g)
    if n < 1e-8:
        return 0.0, 0.0, -1.0
    g /= n
    return float(g[0]), float(g[1]), float(g[2])

# ============================================================
# 2) Synthetic Dataset (RGBOnlyDataset)
# ============================================================

class RGBOnlyDataset(Dataset):
    """
    Synthetic:
      SYNTH_ROOT/split/render_X/
        frame_001.png .. frame_125.png
        render_metadata.json (rotation_over_time: {frame: {x,y,z}})
    """
    def __init__(self, root_dir, split='train', transform=None):
        self.root_dir = os.path.join(root_dir, split)
        self.transform = transform
        self.samples = []  # (frame_path, (gx,gy,gz))

        if not os.path.exists(self.root_dir):
            print(f"‚ö†Ô∏è Synthetic path not found: {self.root_dir}")
            return

        for fld in sorted(os.listdir(self.root_dir)):
            render_path = os.path.join(self.root_dir, fld)
            js = os.path.join(render_path, "render_metadata.json")
            if not os.path.exists(js):
                continue

            try:
                with open(js, "r") as f:
                    meta = json.load(f)
                rot = meta.get("rotation_over_time", {})

                for i in range(1, 126):
                    fp = os.path.join(render_path, f"frame_{i:03d}.png")
                    if not os.path.exists(fp) or str(i) not in rot:
                        continue
                    r = rot[str(i)]
                    if not all(k in r for k in ("x","y","z")):
                        continue

                    ex, ey, ez = float(r["x"]), float(r["y"]), float(r["z"])
                    gx, gy, gz = new_euler_to_gaze_vector(ex, ey, ez)
                    self.samples.append((fp, (gx, gy, gz)))
            except Exception as e:
                print(f"Parse error in {render_path}: {e}")

        print(f"SYNTH-{split}: {len(self.samples)} samples")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        fp, gaze = self.samples[idx]
        img = Image.open(fp).convert("RGB")
        if self.transform is not None:
            img = self.transform(img)
        tgt = torch.tensor(gaze, dtype=torch.float32)
        return img, tgt

# ============================================================
# 3) REAL TEyeD Dataset
# ============================================================

class TEyeDRealDataset(Dataset):
    """
    REAL TEyeD:
      REAL_ROOT/split/*.mp4
      REAL_ROOT/split/*.mp4gaze_vec.txt
    TXT satƒ±r formatƒ±:
      FRAME;X;Y;Z;
      1;-0.0218;-0.0039;0.9997;
      ...
    """
    def __init__(self, root_dir, split="train", transform=None):
        self.root_dir = os.path.join(root_dir, split)
        self.transform = transform
        self.samples = []  # (video_path, frame_idx, (gx,gy,gz))

        if not os.path.exists(self.root_dir):
            print(f"‚ö†Ô∏è REAL path not found: {self.root_dir}")
            return

        video_files = sorted([f for f in os.listdir(self.root_dir) if f.lower().endswith(".mp4")])
        print(f"[REAL-{split}] found {len(video_files)} videos.")

        for vname in tqdm(video_files, desc=f"Indexing REAL-{split}"):
            vpath = os.path.join(self.root_dir, vname)
            txt_path = vpath + "gaze_vec.txt"
            if not os.path.exists(txt_path):
                print(f"‚ö†Ô∏è Gaze file missing for {vname}")
                continue

            frame_ids, gazes = self._load_gaze_file(txt_path)

            cap = cv2.VideoCapture(vpath)
            if not cap.isOpened():
                print(f"‚ö†Ô∏è Could not open video: {vpath}")
                continue
            total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            cap.release()

            L = min(total_frames, len(frame_ids))
            for i in range(L):
                self.samples.append((vpath, int(frame_ids[i]), gazes[i]))

        print(f"REAL-{split}: {len(self.samples)} samples")

    def _load_gaze_file(self, txt_path):
        frame_ids = []
        gazes = []
        with open(txt_path, "r") as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue
                if line.upper().startswith("FRAME"):
                    continue
                parts = line.split(";")
                if len(parts) < 4:
                    continue
                try:
                    fid = int(float(parts[0]))
                    gx = float(parts[1]); gy = float(parts[2]); gz = float(parts[3])
                except ValueError:
                    continue
                frame_ids.append(max(0, fid - 1))  # 1-based -> 0-based
                gazes.append([gx, gy, gz])
        frame_ids = np.array(frame_ids, dtype=np.int64)
        gazes = np.array(gazes, dtype=np.float32)
        return frame_ids, gazes

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        vpath, fidx, gaze = self.samples[idx]
        cap = cv2.VideoCapture(vpath)
        if not cap.isOpened():
            raise RuntimeError(f"Could not open video {vpath}")
        cap.set(cv2.CAP_PROP_POS_FRAMES, fidx)
        ok, frame = cap.read()
        cap.release()
        if not ok:
            raise RuntimeError(f"Could not read frame {fidx} from {vpath}")
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        if self.transform is not None:
            frame = self.transform(frame)

        tgt = torch.tensor(gaze, dtype=torch.float32)
        return frame, tgt

# ============================================================
# 4) Transforms (Synthetic: PIL, Real: numpy -> PIL)
# ============================================================

IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]

synth_tf_train = T.Compose([
    T.Grayscale(num_output_channels=3),
    T.Resize((224, 224)),
    T.ColorJitter(brightness=0.2, contrast=0.2),
    T.ToTensor(),
    T.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

synth_tf_eval = T.Compose([
    T.Grayscale(num_output_channels=3),
    T.Resize((224, 224)),
    T.ToTensor(),
    T.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

real_tf_train = T.Compose([
    T.ToPILImage(),
    T.Resize((224, 224)),
    T.ColorJitter(brightness=0.2, contrast=0.2),
    T.ToTensor(),
    T.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

real_tf_eval = T.Compose([
    T.ToPILImage(),
    T.Resize((224, 224)),
    T.ToTensor(),
    T.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

# ============================================================
# 5) Datasets & DataLoaders
# ============================================================

# Synthetic
synth_train_dataset = RGBOnlyDataset(SYNTH_ROOT, 'train', synth_tf_train)
synth_val_dataset   = RGBOnlyDataset(SYNTH_ROOT, 'val',   synth_tf_eval)
synth_test_dataset  = RGBOnlyDataset(SYNTH_ROOT, 'test',  synth_tf_eval)

# REAL
real_train_dataset = TEyeDRealDataset(REAL_ROOT, 'train', real_tf_train)
real_val_dataset   = TEyeDRealDataset(REAL_ROOT, 'val',   real_tf_eval)
real_test_dataset  = TEyeDRealDataset(REAL_ROOT, 'test',  real_tf_eval)

BATCH_SIZE = 8
NUM_WORKERS = 4 if device=="cuda" else 2

def make_loader(ds, shuffle):
    return DataLoader(
        ds,
        batch_size=BATCH_SIZE,
        shuffle=shuffle,
        num_workers=NUM_WORKERS,
        pin_memory=(device=="cuda"),
        persistent_workers=(device=="cuda" and NUM_WORKERS>0 and len(ds)>0),
        prefetch_factor=4 if device=="cuda" else 2
    )

synth_train_loader = make_loader(synth_train_dataset, True)
synth_val_loader   = make_loader(synth_val_dataset,   False)
synth_test_loader  = make_loader(synth_test_dataset,  False)

real_train_loader = make_loader(real_train_dataset, True)
real_val_loader   = make_loader(real_val_dataset,   False)
real_test_loader  = make_loader(real_test_dataset,  False)

print(f"SYNTH Train: {len(synth_train_dataset)} | Val: {len(synth_val_dataset)} | Test: {len(synth_test_dataset)}")
print(f"REAL  Train: {len(real_train_dataset)} | Val: {len(real_val_dataset)} | Test: {len(real_test_dataset)}")

# ============================================================
# 6) Model: ResNet50 -> 3D gaze vector
# ============================================================

gaze_model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
gaze_model.fc = nn.Linear(gaze_model.fc.in_features, 3)
gaze_model = gaze_model.to(device)

# Partial load from synthetic checkpoint (if format matches)
try:
    ckpt = torch.load(CKPT_PATH, map_location="cpu", weights_only=False)
    sd = ckpt.get("state_dict", ckpt)
    rep = gaze_model.load_state_dict(sd, strict=False)
    print("Partial load from last.pt done.")
    print(f"Missing keys: {len(rep.missing_keys)} | Unexpected keys: {len(rep.unexpected_keys)}")
except Exception as e:
    print(f"Checkpoint load skipped: {e}")

# ============================================================
# 7) Loss & Metrics (normalized vectors)
# ============================================================

def combined_loss(pred, tgt, alpha=0.7):
    """
    alpha * (1 - cos) + (1 - alpha) * MSE, normalize edilmi≈ü vekt√∂rler √ºst√ºnde.
    """
    pred_n = F.normalize(pred, dim=1)
    tgt_n  = F.normalize(tgt,  dim=1)
    cos = torch.sum(pred_n * tgt_n, dim=1).clamp(-1.0, 1.0)
    cos_loss = (1.0 - cos).mean()
    mse_loss = F.mse_loss(pred_n, tgt_n)
    return alpha * cos_loss + (1.0 - alpha) * mse_loss

def angular_error_deg(pred, gt):
    pred = F.normalize(pred, dim=1)
    gt   = F.normalize(gt,   dim=1)
    cos_sim = torch.clamp(torch.sum(pred * gt, dim=1), -1.0, 1.0)
    return torch.rad2deg(torch.acos(cos_sim))

# ============================================================
# 8) Optimizer: differential LR + warmup + scheduler
# ============================================================

head_params = list(gaze_model.fc.parameters())
backbone_params = [p for n,p in gaze_model.named_parameters() if not n.startswith('fc.')]

optimizer = torch.optim.AdamW(
    [{'params': backbone_params, 'lr': 1e-5},
     {'params': head_params,     'lr': 5e-5}],
    weight_decay=1e-4
)

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', factor=0.5, patience=3
)

WARMUP_EPOCHS = 5
BASE_LR_BB, BASE_LR_HD = 1e-5, 5e-5

def apply_warmup(opt, epoch):
    if epoch <= WARMUP_EPOCHS:
        k = epoch / float(WARMUP_EPOCHS)  # 0.2..1.0 arasƒ±
        opt.param_groups[0]['lr'] = BASE_LR_BB * k
        opt.param_groups[1]['lr'] = BASE_LR_HD * k

# ============================================================
# 9) Mixed Train / Val Loop (REAL 90% + SYNTH 10%)
# ============================================================

EPOCHS = 20
REAL_WEIGHT  = 0.5
SYNTH_WEIGHT = 0.5

USE_SYNTH = len(synth_train_dataset) > 0
if not USE_SYNTH:
    print("‚ö†Ô∏è Synthetic train dataset bo≈ü; training only on REAL.")
    REAL_WEIGHT, SYNTH_WEIGHT = 1.0, 0.0

best_val = float("inf")
train_losses, val_losses, val_angles, bb_lrs, hd_lrs = [], [], [], [], []

MAX_REAL_BATCHES_PER_EPOCH = 500   # istersen arttƒ±r/azalt
MAX_VAL_BATCHES            = 100

for epoch in range(1, EPOCHS+1):
    gaze_model.train()
    apply_warmup(optimizer, epoch)

    tr_loss = 0.0
    used_batches = 0
    synth_iter = iter(synth_train_loader)

    # --------- Train (REAL + optional SYNTH) ----------
    for batch_idx, (xr, tr) in enumerate(
        tqdm(real_train_loader, desc=f"Epoch {epoch}/{EPOCHS} [Train]", leave=False)
    ):
        if batch_idx >= MAX_REAL_BATCHES_PER_EPOCH:
            break

        xr, tr = xr.to(device), tr.to(device)
        optimizer.zero_grad(set_to_none=True)

        out_r = gaze_model(xr)
        loss_r = combined_loss(out_r, tr, alpha=0.7)

        if USE_SYNTH and SYNTH_WEIGHT > 0 and len(synth_train_dataset) > 0:
            try:
                xs, ts = next(synth_iter)
            except StopIteration:
                synth_iter = iter(synth_train_loader)
                xs, ts = next(synth_iter)
            xs, ts = xs.to(device), ts.to(device)
            out_s = gaze_model(xs)
            loss_s = combined_loss(out_s, ts, alpha=0.7)

            loss = REAL_WEIGHT * loss_r + SYNTH_WEIGHT * loss_s
        else:
            loss = loss_r

        loss.backward()
        optimizer.step()

        tr_loss += loss.item()
        used_batches += 1

    tr_loss /= max(1, used_batches)

    # --------- Validation (REAL only) ----------
    gaze_model.eval()
    va_loss = 0.0
    va_ang_list = []

    with torch.no_grad():
        for batch_idx, (xr, tr) in enumerate(
            tqdm(real_val_loader, desc=f"Epoch {epoch}/{EPOCHS} [Val]", leave=False)
        ):
            if batch_idx >= MAX_VAL_BATCHES:
                break

            xr, tr = xr.to(device), tr.to(device)
            out = gaze_model(xr)
            loss = combined_loss(out, tr, alpha=0.7)
            va_loss += loss.item()
            va_ang_list.append(angular_error_deg(out, tr).mean().item())

    va_loss /= max(1, min(MAX_VAL_BATCHES, len(real_val_loader)))
    va_ang = float(np.mean(va_ang_list)) if len(va_ang_list) else float('inf')

    train_losses.append(tr_loss)
    val_losses.append(va_loss)
    val_angles.append(va_ang)
    bb_lrs.append(optimizer.param_groups[0]['lr'])
    hd_lrs.append(optimizer.param_groups[1]['lr'])

    print(f"Epoch {epoch:02d} | Train Loss: {tr_loss:.6f} | Val Loss: {va_loss:.6f} "
          f"| Val Angular(REAL): {va_ang:.2f}¬∞ | LR(bb,hd)=({bb_lrs[-1]:.6e},{hd_lrs[-1]:.6e})")

    prev_bb, prev_hd = optimizer.param_groups[0]['lr'], optimizer.param_groups[1]['lr']
    scheduler.step(va_loss)
    if optimizer.param_groups[0]['lr'] < prev_bb or optimizer.param_groups[1]['lr'] < prev_hd:
        print(f"üîª LR reduced to backbone={optimizer.param_groups[0]['lr']:.6e}, head={optimizer.param_groups[1]['lr']:.6e}")

    if va_loss + 1e-12 < best_val:
        best_val = va_loss
        torch.save(gaze_model.state_dict(), SAVE_BEST)
        print("‚úÖ Best mixed model saved:", SAVE_BEST)

print("Best REAL val loss:", best_val)

# ============================================================
# 10) Test (REAL + optional SYNTH)
# ============================================================

gaze_model.load_state_dict(torch.load(SAVE_BEST, map_location=device))
gaze_model.to(device)
gaze_model.eval()

# REAL Test
test_angles_real = []
with torch.no_grad():
    for xr, tr in tqdm(real_test_loader, desc="REAL Test"):
        xr, tr = xr.to(device), tr.to(device)
        ang = angular_error_deg(gaze_model(xr), tr)
        test_angles_real.append(ang.mean().item())
print(f"‚úÖ REAL TEST Angular Error: {np.mean(test_angles_real):.2f}¬∞")

# SYNTH Test (opsiyonel)
if len(synth_test_dataset) > 0:
    test_angles_synth = []
    with torch.no_grad():
        for xs, ts in tqdm(synth_test_loader, desc="SYNTH Test"):
            xs, ts = xs.to(device), ts.to(device)
            ang = angular_error_deg(gaze_model(xs), ts)
            test_angles_synth.append(ang.mean().item())
    print(f"‚úÖ SYNTH TEST Angular Error: {np.mean(test_angles_synth):.2f}¬∞")
else:
    print("‚ö†Ô∏è SYNTH TEST dataset bo≈ü, SYNTH test atlandƒ±.")

# ============================================================
# 11) Plot'lar
# ============================================================

fig, ax1 = plt.subplots(figsize=(9,5))
ax1.plot(train_losses, label="Train Loss (mixed)")
ax1.plot(val_losses,   label="Val Loss (REAL)")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.grid(True)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.plot(val_angles, label="Val Angular (REAL, ¬∞)", linestyle='--')
ax2.set_ylabel("Angular Error (¬∞)")
ax2.legend(loc='upper right')

plt.title("Training Progress (Mixed Synthetic + REAL)")
plt.show()

plt.figure(figsize=(9,4))
plt.plot(bb_lrs, label="LR Backbone")
plt.plot(hd_lrs, label="LR Head")
plt.title("Learning Rate Schedule")
plt.xlabel("Epoch")
plt.ylabel("LR")
plt.grid(True)
plt.legend()
plt.show()
```
